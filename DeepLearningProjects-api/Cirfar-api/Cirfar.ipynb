```python
train_img=train_img/255.0
test_img=test_img/255.0
```

```python
class_names=['airport','automobile','bird','cat','deer','dog',
            'frog','horse','ship','truck']
```

```python
cnn=models.Sequential()

cnn.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)))
cnn.add(layers.MaxPooling2D(2,2))

cnn.add(layers.Conv2D(32,(3,3),activation='relu'))
cnn.add(layers.MaxPooling2D(2,2))

cnn.add(layers.Conv2D(32,(3,3),activation='relu'))
cnn.add(layers.MaxPooling2D(2,2))


cnn.add(layers.Flatten())
cnn.add(layers.Dense(100,activation='relu'))
cnn.add(layers.Dense(80,activation='relu'))
cnn.add(layers.Dense(40,activation='relu'))
cnn.add(layers.Dense(10,activation='softmax'))
```

```python
cnn.summary()
```

```python
cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history=cnn.fit(train_img ,train_lbls,epochs=1)
```

```python
cnn.evaluate(test_img,test_lbl)
```
